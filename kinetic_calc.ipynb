```python

# Ничего не загружаем. Просто меняете DATA и C0, потом запускаете ячейку.

import re
import pandas as pd
from IPython.display import display

#  настройки формулы
INTERCEPT = 0.00716
SLOPE = 17257.3404
VOLUME_ML = 3.0

# C0 нужен только для процента высвобождения.
# Если процент не нужен, поставьте: C0 = None

C0 = 0.0009759148678 #@param {type:"number"}

# Вставляйте сюда 2 столбца: время и оптическая плотность.
# Можно с пробелами, табами или точкой с запятой.
# Десятичная запятая поддерживается.
DATA = """
1 0,192350
5 0,170120
10 0,208790
15 0,279450
20 0,308960
30 0,379120
45 0,363280
60 0,380040
90 0,374980
120 0,361130
180 0,386540
240 0,382900
300 0,415470
1800 0,03302
2900 0,06316
4000 0,06976
5100 0,06984
6200 0,06941
"""


def parse_number(x):
    return float(str(x).strip().replace(',', '.'))


def read_data(text):
    rows = []
    skipped = []

    for line in text.strip().splitlines():
        line = line.strip()
        if not line or line.startswith('#'):
            continue

        parts = re.split(r"[\t;\s]+", line)
        parts = [p for p in parts if p]

        if len(parts) < 2:
            skipped.append(line)
            continue

        try:
            time_value = parse_number(parts[0])
            absorbance = parse_number(parts[1])
            rows.append((time_value, absorbance))
        except ValueError:
            skipped.append(line)

    if not rows:
        raise ValueError("Нет ни одной строки с двумя числами: время и оптическая плотность.")

    if skipped:
        print(f"Пропущено строк: {len(skipped)}")

    return rows


rows = read_data(DATA)

df = pd.DataFrame(rows, columns=["Время, мин", "Опт. плотность"])

amount_col = f"C в {VOLUME_ML:g} мл, мг"
df["C, мг/мл"] = (df["Опт. плотность"] + INTERCEPT) / SLOPE
df[amount_col] = df["C, мг/мл"] * VOLUME_ML
df["Сумма C, мг"] = df[amount_col].cumsum()

if C0 is not None:
    df["Процент высвобождения"] = df["Сумма C, мг"] / C0 * 100


pd.set_option('display.max_rows', 200)
pd.set_option('display.max_columns', 20)
pd.set_option('display.float_format', lambda x: f'{x:.16f}')
display(df)


output_filename = "raschet_reliza.xlsx"
df.to_excel(output_filename, index=False)
print(f"Готово. Файл сохранён: {output_filename}")


try:
    from google.colab import files
    files.download(output_filename)
except Exception:
    pass
```
